#주제: 별자리 챗봇

1. 사용자에게 생일 물어보기
2. 생일에 해당하는 별자리 알려주기
3. 별자리 성격 / 특징 알려주기 (3문장 이내)
4. 응답에는 제약이 없음

```python
from PIL import Image
import matplotlib.pyplot as plt

img1 = Image.open("res/Quiz1.png")
img2 = Image.open("res/Quiz2.png")

# 1행 2열 subplot으로 표시
fig, axes = plt.subplots(2, 1, figsize=(10, 10))

axes[0].imshow(img1)
axes[0].set_title("Quiz")
axes[0].axis("off")

axes[1].imshow(img2)
axes[1].set_title("Result")
axes[1].axis("off")

plt.show()
```
