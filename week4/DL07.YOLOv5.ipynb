## YOLOv5 학습 with COCO128

**📦 COCO128 데이터셋 자동 다운로드**

- YOLOv5는 coco128.yaml 설정파일과 함께 데이터를 자동으로 다운로드합니다.

**🚀 학습 실행**
- `--img 640`: 이미지 리사이즈 크기

- `--batch 16`: 배치 크기

- `--epochs 10`: 에폭 수 (더 늘릴 수도 있어요)

- `--data data/coco128.yaml`: 설정 파일 (자동 포함됨)

- `--weights yolov5s.pt`: 사전학습된 모델

- `--name`: 실험 이름



````python
!python train.py \
  --img 640 \
  --batch 16 \
  --epochs 10 \
  --data data/coco128.yaml \
  --weights yolov5s.pt \
  --name coco128_yolo
````

## YAML 파일 구조 예시

✅ YOLOv5용 커스텀 데이터셋 폴더 구성 예시
```yaml
dataset/
├── images/
│   ├── train/
│   │   ├── img001.jpg
│   │   ├── img002.jpg
│   │   └── ...
│   └── val/
│       ├── img101.jpg
│       ├── img102.jpg
│       └── ...
├── labels/
│   ├── train/
│   │   ├── img001.txt
│   │   ├── img002.txt
│   │   └── ...
│   └── val/
│       ├── img101.txt
│       ├── img102.txt
│       └── ...
└── data.yaml

```
- images/train/, images/val/: 이미지(.jpg/.png)

- labels/train/, labels/val/: 이미지와 동일한 이름의 라벨(.txt) 파일

- data.yaml: 경로 및 클래스 정의

✅ 라벨 파일 (labels/train/img001.txt) 예시
```
<class_id> <x_center> <y_center> <width> <height>
```
```
0 0.5 0.5 0.4 0.4     ← 사과 (apple), 가운데 위치
1 0.7 0.3 0.2 0.3     ← 바나나 (banana), 우상단
```
- 0: apple

- 1: banana

- 2: orange


```yaml
# data/my_custom.yaml

train: /content/dataset/images/train
val: /content/dataset/images/val

nc: 3
names: ['apple', 'banana', 'orange']
```

````python
import yaml

yaml_data = {
    'train': '/content/dataset/images/train',
    'val': '/content/dataset/images/val',
    'nc': 3,
    'names': ['apple', 'banana', 'orange']
}

with open('data/my_custom.yaml', 'w') as f:
    yaml.dump(yaml_data, f)
````
