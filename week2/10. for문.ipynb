## for 문의 기본 구조

```
for 변수 in 리스트(또는 튜플, 문자열):
    수행할_문장1
    수행할_문장2
    ...
```

### 1. 전형적인 for 문

````python
test_list = ['one', 'two', 'three'] 
for i in test_list: 
    print(i)
````

### 2. 다양한 for 문의 사용

````python
a = [(1,2), (3,4), (5,6)]
for (first, last) in a:
    print(first + last)
````

### 3. for 문의 응용
```
총 5명의 학생이 시험을 보았는데 시험 점수가 60점 이상이면 합격이고 그렇지 않으면 불합격이다. 합격인지, 불합격인지 결과를 보여 주시오.
```


````python
marks = [90, 25, 67, 45, 80]   # 학생들의 시험 점수 리스트

number = 0   # 학생에게 붙여 줄 번호
for mark in marks:   # 90, 25, 67, 45, 80을 순서대로 mark에 대입
    number = number +1 
    if mark >= 60: 
        print("%d번 학생은 합격입니다." % number)
    else: 
        print("%d번 학생은 불합격입니다." % number)
````

### for 문과 continue 문

````python
marks = [90, 25, 67, 45, 80]

number = 0 
for mark in marks: 
    number = number +1 
    if mark < 60:
        continue 
    print("%d번 학생 축하합니다. 합격입니다. " % number)
````

### for 문과 함께 자주 사용하는 range 함수

````python
a = range(10)
a
````

````python
a = range(1, 11)    # 1부터 10까지의 숫자 생성(1이상 11미만)
a
````

````python
add = 0 
for i in range(1, 11): 
    add = add + i

add
````

````python
marks = [90, 25, 67, 45, 80]
for number in range(len(marks)):
    if marks[number] < 60: 
        continue
    print("%d번 학생 축하합니다. 합격입니다." % (number+1))
````

### for와 range를 이용한 구구단

````python
for i in range(2,10):        # 1번 for문
    for j in range(1, 10):   # 2번 for문
        print(i*j, end=" ") 
    print('')
````

### 리스트 컴프리헨션 사용하기
```
[표현식 for 항목 in 반복_가능_객체 if 조건문]
```

```
[표현식 for 항목1 in 반복_가능_객체1 if 조건문1
      for 항목2 in 반복_가능_객체2 if 조건문2
      ...
      for 항목n in 반복_가능_객체n if 조건문n]
```

````python
a = [1,2,3,4]
result = []
for num in a:
    result.append(num*3)
result
````

````python
a = [1,2,3,4]
result = [num * 3 for num in a]
result
````

````python
a = [1,2,3,4]
result = [num * 3 for num in a if num % 2 == 0]
result
````

````python
result = [x*y for x in range(2,10)
              for y in range(1,10)]
result
````

### for 문과 break 문

````python
for i in range(10):
    if i == 5:
        break
    print(i)
````

### for-else 문

````python
for i in range(5):
    print(i)
else:
    print("for 문이 정상 종료되었습니다.")
````

````python
for i in range(5):
    if i == 3:
        break
    print(i)
else:
    print("for 문이 정상 종료되었습니다.")
````

### enumerate 함수 활용하기

````python
fruits = ['apple', 'banana', 'orange']
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
````

````python
fruits = ['apple', 'banana', 'orange']
for i, fruit in enumerate(fruits, 1):  # 1부터 시작
    print(f"{i}: {fruit}")
````

### zip 함수로 여러 리스트 함께 순회하기

````python
names = ['홍길동', '김철수', '이영희']
scores = [85, 92, 78]
for name, score in zip(names, scores):
    print(f"{name}: {score}점")
````

````python
names = ['홍길동', '김철수', '이영희']
korean = [85, 92, 78]
english = [90, 88, 95]
for name, kor, eng in zip(names, korean, english):
    print(f"{name}: 국어 {kor}점, 영어 {eng}점")
````
